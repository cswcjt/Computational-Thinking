```python
def fastFib(n, memo = {}): # memo : 앞으로 fib의 결과를 담을 딕셔너리
    """Assumes n is an int >= 0, memo used only by recursive calls
       Returns Fibonacci of n"""
    if n == 0 or n == 1:
        return 1
    try:
        return memo[n]
    except KeyError:
        #print(memo)
        result = fastFib(n-1, memo) + fastFib(n-2, memo)
        #print(result)
        memo[n] = result
        return result

for i in range(121):
    print('fib(' + str(i) + ') =', fastFib(i))
```
